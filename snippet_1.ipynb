## Построение графиков с помощью `arange` и `linspace`

```python
x = np.arange(-2, 2.1, 0.1)
y = [v**3 for v in x]

plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-2, 2, 100)
y = [v**3 for v in x]

plt.grid()
plt.plot(x, y)
```
