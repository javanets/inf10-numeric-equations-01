## Построение графиков с помощью `arange` и `linspace`

```python
x = np.arange(-2, 2.1, 0.1)
y = [v**3 for v in x]

plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-2, 2, 100)
y = [v**3 for v in x]

plt.grid()
plt.plot(x, y)
```

## Какой метод выбрать: перебор или бисекция?

* **Метод перебора**:
  * можно применять для нахождения всех решений $f(x) = 0$ на $[a;b]$
  * вычислительно неэффективен при малой полосе разбиения $\delta$.

* **Метод бисекции**:
  * можно применять только для $f(x)$, имеющей на $[a;b]$ **единственный** корень
  * вычислительно эффективен
